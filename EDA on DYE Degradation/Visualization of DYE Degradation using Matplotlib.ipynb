```python
x = [10,20,30,40,50]
ya = [24.44,35.55,36.66,37.77,50]
y = [84.89,83.54,87.05,89.20,92.80]
```

```python
from matplotlib import pyplot as plt
%matplotlib inline
plt.plot(x,ya,label="Dye Degrade (Cacl2)",marker='D',linestyle='--')
plt.plot(x,y,label="Dye Degrade(5N- H2So4)",marker='D',linestyle='--')
#plt.plot(dev_x,py_dev_y,label="python")
plt.xlabel('Time')
plt.ylabel('% of removal')
plt.legend()
plt.grid()
plt.title("Effect of Contact Time  ")
plt.show()
```
