# univariate, Bivariate and Multivariate Analysis

```python
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv("IRIS.csv")
```

```python
df
```

```python
df.shape
```

## Univariate Analysis

```python
setosa = df.loc[df['species']=='Iris-setosa']
virginica = df.loc[df['species']=='Iris-virginica']
versicolor = df.loc[df['species']=='Iris-versicolor']
```

```python
plt.plot(setosa['sepal_length'],np.zeros_like(setosa['sepal_length']),'o')
plt.plot(virginica['sepal_length'],np.zeros_like(setosa['sepal_length']),'o')
plt.plot(versicolor['sepal_length'],np.zeros_like(setosa['sepal_length']),'o')
plt.xlabel('petal length')
```

# Bivariate analysis

```python
sns.FacetGrid(df,hue='species',height=5).map(plt.scatter,"petal_length","sepal_width")
```

```python
sns.FacetGrid(df,hue='species',height=5).map(plt.scatter,"sepal_length","sepal_width")
```
